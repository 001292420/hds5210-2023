### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
# Here first i imported pandas library and my datafile then i printed the number of rows and columns in my data file.
import pandas as pd
file_path = '/content/VEHSS_Composite_Prevalence_Estimates.csv'
data = pd.read_csv(file_path)
num_rows, num_columns = data.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
#Here i printed the number of unique values in each column and column with most unique values.
unique_values_per_column = data.nunique()
print("Number of unique values in each column:")
print(unique_values_per_column)
max_unique_column = unique_values_per_column.idxmax()
max_unique_count = unique_values_per_column.max()
print(f"\nColumn with the most unique values:")
print(f"Column: {max_unique_column}")
print(f"Number of unique values: {max_unique_count}")
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
#Here i identified the unvarying columns. columns where the value for every row is same.
unvarying_columns = data.columns[data.nunique() == 1]
print("Columns where the value for every row is the same:")
print(unvarying_columns)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
#Here i created a histogram that shows Distribution of values for Low Confidence limit. Most of the low confidence limit values lies between 0-5.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/VEHSS_Composite_Prevalence_Estimates.csv', low_memory=False)
plt.hist(data['Low_Confidence_limit'], bins=10, edgecolor='black')
plt.xlabel('Low_Confidence_limit')
plt.ylabel('Frequency')
plt.title('Distribution of values for Low Confidence limit')
plt.show()









```
